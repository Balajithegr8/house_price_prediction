```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
lab = df['price']
conv_dates = [1 if values == 2014 else 0 for values in df.date ]
df['date'] = conv_dates
train_data = df.drop(['id', 'price'],axis=1)
print(train_data)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(train_data , lab , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
pickle.dump(reg, open('model.pkl','wb'))
```

```python
reg.score(x_test,y_test)
```

Hmm,accuracy seems low ,less than 85% is not considered real life applicable. let's apply some gradient descent

```python
from sklearn import ensemble
cf=ensemble.GradientBoostingRegressor(n_estimators=400,max_depth=5,min_samples_split=2,learning_rate=0.1,loss='squared_error')
```

```python
cf.fit(x_train,y_train)
```

```python
cf.score(x_test,y_test)
```

Now this is acceptable 92%

```python
cf=ensemble.GradientBoostingRegressor(n_estimators=1000,max_depth=5,min_samples_split=2,learning_rate=0.07,loss='huber')
```

```python
cf.fit(x_train,y_train)
```

```python
cf.score(x_test,y_test)
```
